```python
import pandas as pd
df = pd.read_csv("C:\\Users\\User\\Downloads\\train.csv")
print(df.head())
```

This will give us an overview of the columns and the type of data present in the dataset.

```python
# Check for missing values
print(df.isnull().sum())

# Handle missing values
median_age = df['Age'].median()
df['Age'].fillna(median_age, inplace=True)

# For the 'Embarked' column, we can fill missing values with the mode (most frequent value)
mode_embarked = df['Embarked'].mode()[0]
df['Embarked'].fillna(mode_embarked, inplace=True)

```

Now that missing values are handled, Then exploring the relationships between variables and identifying patterns and trends.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for seaborn
sns.set(style="whitegrid")

# Plot the distribution of passenger ages
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Plot the count of passengers by sex
plt.figure(figsize=(8, 5))
sns.countplot(x='Sex', data=df, palette='Set2')
plt.title('Count of Passengers by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()

# Plot the count of passengers by passenger class
plt.figure(figsize=(8, 5))
sns.countplot(x='Pclass', data=df, palette='Set1')
plt.title('Count of Passengers by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

These visualizations provide insights into the distribution of passenger ages, the count of passengers by sex, and the count of passengers by passenger class.



Next, explore the relationship between variables, such as survival rate by sex and passenger class:

```python
# Plot survival rate by sex
plt.figure(figsize=(8, 5))
sns.barplot(x='Sex', y='Survived', data=df, palette='Pastel1')
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()

# Plot survival rate by passenger class
plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', data=df, palette='Pastel2')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

These visualizations help us understand the relationship between survival and other variables like sex and passenger class.

Exploratory data analysis helps in understanding the dataset, identifying patterns, and making informed decisions when building predictive models or deriving insights.
